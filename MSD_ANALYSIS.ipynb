```python
import pandas as pd
```

```python
deliveries=pd.read_csv("deliveries.csv")
deliveries
```

```python
unique_matches=[]
for id_match in msd_unique_match_list:
    msd_df=matches[matches["id"]==id_match]
    unique_matches.append(msd_df.iloc[0])
unique_matches
```

```python
unique_df=pd.DataFrame(unique_matches)
unique_df
```

```python
unique_df['date']=pd.to_datetime(unique_df['date'])
```

```python
unique_df
```

```python
unique_df=unique_df.sort_values(by="date")
unique_df
```

```python
runs=[]
for index,row in unique_df.iterrows():
    match_id=row["id"]
    msd_temp=deliveries[(deliveries["match_id"]==match_id) & (deliveries['batsman']=="MS Dhoni")]
    runs_msd=msd_temp['batsman_runs'].sum()
    runs.append(runs_msd)
runs
```

```python
unique_df["runs"]=runs
```

```python
unique_df
```

```python
balls=[]
sr=[]
ten=[]
ten_thirty=[]
thirty_plus=[]
for index,row in unique_df.iterrows():
    match_id=row["id"]
    msd_temp=deliveries[(deliveries["match_id"]==match_id) & (deliveries['batsman']=="MS Dhoni")]
    msd_temp2=msd_temp[msd_temp["wide_runs"]==0]
    msd_balls=len(msd_temp[msd_temp["wide_runs"]==0])
    balls.append(msd_balls)
    if(msd_balls<=10):
        ten.append(msd_temp["batsman_runs"].sum())
        ten_thirty.append(0)
        thirty_plus.append(0)
    elif((msd_balls>10) & (msd_balls<=30)):
        ten.append(msd_temp2.iloc[0:10]['batsman_runs'].sum())
        ten_thirty.append(msd_temp2.iloc[10:]['batsman_runs'].sum())
        thirty_plus.append(0)
    else:
        ten.append(msd_temp2.iloc[0:10]['batsman_runs'].sum())
        ten_thirty.append(msd_temp2.iloc[10:30]['batsman_runs'].sum())
        thirty_plus.append(msd_temp2.iloc[30:]['batsman_runs'].sum())
```

```python
balls
```

```python
ten
```

```python
ten_thirty
```

```python
thirty_plus
```

```python
unique_df['balls']=balls
unique_df['ten']=ten
unique_df['ten_thirty']=ten_thirty
unique_df['thirty_plus']=thirty_plus
unique_df
```

```python
sr=[]
for index,row in unique_df.iterrows():
    sr_temp=(row['runs']/row['balls'])*100
    sr.append(sr_temp)
sr
```

```python
unique_df['sr']=sr
```

```python
unique_df
```

```python
msd_data=msd_group_season.mean()
```

```python
dot_balls=[]
for index,row in unique_df.iterrows():
    match_id=row["id"]
    msd_temp=deliveries[(deliveries["match_id"]==match_id) & (deliveries['batsman']=="MS Dhoni")]
    dictionary=msd_temp['batsman_runs'].value_counts().to_dict()
    if(0 in dictionary.keys()):
        dot_ball=msd_temp['batsman_runs'].value_counts().to_dict()[0]
    else:
        dot_ball=0
    dot_balls.append(dot_ball)
dot_balls
```

```python
unique_df['dot_balls']=dot_balls
unique_df
```

```python
dot_ball_per=[]
for index,row in unique_df.iterrows():
    dot_ball_temp=(row["dot_balls"]/row["balls"])*100
    dot_ball_per.append(dot_ball_temp)
dot_ball_per
```

```python
unique_df['dot_ball_per']=dot_ball_per
unique_df
```

```python
boundaries_per=[]
for index,row in unique_df.iterrows():
    match_id=row["id"]
    msd_temp=deliveries[(deliveries["match_id"]==match_id) & (deliveries['batsman']=="MS Dhoni")]
    dictionary=msd_temp['batsman_runs'].value_counts().to_dict()
    boundary=0
    if(4 in dictionary.keys() and 6 in dictionary.keys()):
        boundary+=(dictionary[4])*4 + (dictionary[6])*6
    elif(6 in dictionary.keys()):
        boundary+=(dictionary[6])*6
    elif(4 in dictionary.keys()):
        boundary+=(dictionary[4])*4
    else:
        boundary=0
    if(row['runs']!=0):
        boundaries_per_temp=(boundary/row['runs'])*100
    else:
        boundaries_per_temp=0
    boundaries_per.append(boundaries_per_temp)
boundaries_per
```

```python
unique_df['boundaries_per']=boundaries_per
```
