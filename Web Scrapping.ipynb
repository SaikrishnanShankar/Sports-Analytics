```python
all_players=[]
for k in range(1, 62):
    url="https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;page="+str(k)+";template=results;type=batting"
    text=urlopen(url)
    soup=BeautifulSoup(text, "lxml")
    table=soup.findAll('table',attrs={'class':'engineTable'})
    int_table=table[0]
    for temp_table in table:
        caption_tag=temp_table.findAll('caption')
        if(len(caption_tag)>0):
            int_table=temp_table
    tr_list=int_table.findAll('tr',attrs={'class':'data1'})
    for tr in tr_list:
        td_list=tr.findAll('td')
        row_data=[]
        for td in td_list:
            td_str=str(td)
            clean_text=BeautifulSoup(td_str,"lxml").get_text()
            row_data.append(clean_text)
        all_players.append(row_data)
```

```python
all_players
```

```python
all_players_df=pd.DataFrame(all_players)
```
